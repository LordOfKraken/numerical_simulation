The genetic algorithm i used to solve the Trading Salesman Problem describe a possible path between cities as a list of int, each one being the id of the city.

An *individual* is a struct, with a int vector representing the order in which the cities are visited, and two double representing the $L^{(1)}$ and $L^{(2)}$ total travel distance.
A *population* is a vector of individual, that is ordered according to the $L^{(1)}$ or $L^{(2)}$ value, with the individual with the shortest path being at the beginning.

There are four possible mutations that can occur to an indvidual, which are:

- Pair permutation of 2 different cities
- Shifting of all the cities
- Shifting of a group of contiguous cities
- Inversion of the visit order for a group of contiguous cities

and there's also a crossover mechanism, which select a pair of individual and mix the visit visit order between the two. Given a selected mother and father, the crossover operation cut their paths at the same position, conserve the first part of the paths and complete the paths with the missing cities adding them in the order in which they appear in the consort.

To select the individual that undergo a mutation or a crossover i used the suggested selection operatr, that uses the order in which the individual appear in the population. 

The specific parameters used can be passed via the input file, specifically i evolved a population of 100 individual for 1000 generation, with a mutation probability of $10\%$ for each mutation, and a crossover probability of $75\%$.

The city and starting population generation are performed by the program itself, and both are saved in the `data` folder.

I've use the software to find the best path between 30 different cities, placed randomly inside a square, or on a circumference. The results are shown below

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (6*2, 5))
axes = axes.flatten()

# square
gen, l_best, l_ave = np.loadtxt("data/square/fitness.out", unpack = True)
axes[0].plot(gen, l_best, label="Best value $L^{(1)}$")
axes[0].plot(gen, l_ave, label="Average value $< L^{(1)}>$")
axes[0].set_title("Fitness ($< L^{(1)}>$)  - Square")

# circumference
gen, l_best, l_ave = np.loadtxt("data/circumference/fitness.out", unpack = True)
axes[1].plot(gen, l_best, label="Best value $L^{(1)}$")
axes[1].plot(gen, l_ave, label="Average value $< L^{(1)}>$")
axes[1].set_title("Fitness ($< L^{(1)}>$)  - Circumference")


for ax in axes:
    ax.set_ylabel("Fitness")
    ax.set_xlabel("Generation")
    ax.legend()
    ax.grid(True)

fig.tight_layout()
plt.show()
```
