### Exercise 2

In this exercise, i will variationally optimise the ground state of a single quantum particle in a one dimensional (1D) space confined by the following external potential:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

Using the following approximate model for the ground state wave function is the following trial wave function:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

I wrote a Variational Monte Carlo code to find the parameters $\sigma$ and $\mu$ which minimize $\langle {\hat H} \rangle_T$:

$$
\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
mu, sigma, H = np.loadtxt("8.1-2/data/minimize.out", usecols=(1,2,4), unpack = True)

fig = plt.figure(figsize = (9, 6))
ax = Axes3D(fig)
ax.scatter(mu, sigma, H, c=H, marker='.')
ax.set_xlabel('$\mu$')
ax.set_ylabel('$\sigma$')
ax.set_zlabel('$<H>$')
ax.view_init(10, 30)

print("<H> is minimized for mu =", mu[np.argsort(H)[0]], ", sigma =", sigma[np.argsort(H)[0]])
```

The optimal parameters are $\mu = 0.79$ and $\sigma=0.63$. Using these parameters, i ran another simulation and produced an estimate of $\langle {\hat H} \rangle_T$ and its statistical uncertainty, using the data blocking technique. The results obtained are compared with the expected value for comparison

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

plt.figure(1,figsize=(10,5))

x_h, H, err_H = np.loadtxt("8.1-2/data/ave_h.out", unpack = True)
plt.title("Expectation value of Hamiltonian")
plt.xlabel('Number of blocks')
plt.ylabel('$< \hat H >_T$')
plt.axhline(y=-0.46046587969271563, color='r', linestyle="dotted", label = "Numerical solution")
plt.errorbar(x_h, H, err_H, label = "Montecarlo")
plt.grid(True)
plt.legend()
```

```python
# Plot potential and probabilities

def Psi(x,mu,sigma):
    return np.exp(-(x-mu)**2/(2.*sigma**2))+np.exp(-(x+mu)**2/(2.*sigma**2))

plt.figure(2, figsize = (10, 5))

x_sample,y_sample = np.loadtxt("8.1-2/data/histo.out", unpack = True)

scale = 0.35
plt.plot(x_sample, y_sample, label="Sampled wave function")
plt.plot(x,(psi[0])**2, label="Numerical solution", linestyle='dashed')
plt.plot(x, scale*(Psi(x, 0.79, 0.63)**2), label="Analytic curve", color='r', linestyle="dotted")
plt.title("Square modulus wave function")
plt.xlabel("x")
plt.ylabel(r"$|\psi_0(x)|^2$")
plt.xlim(-3,3)
plt.grid(True)
plt.legend()
```

### Exercise 3

With the PIMC/PIGS code provided, i studied the same 1D quantum system, and used the PIGS algorithm ($T=0$) to project a constant trial wave function, $\Psi_T(x)=1$, and the optimized wave function $\Psi_T^{\sigma,\mu}(x)$,  with $\mu = 0.79$ and $\sigma = 0.63$.

The results are compared with those obtained from the VMC code

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (6*2, 5))
axes = axes.flatten()

# constant trial wavefunction
x_pigs, y_pigs, err_pigs = np.loadtxt("8.3/data/const/probability.dat", unpack = True)
axes[0].errorbar(x_pigs, y_pigs, yerr = err_pigs, label = "PIGS")
axes[0].plot(x_sample,y_sample, label="VMC code")
axes[0].set_title("PIGS - Constant wavefunction")

# optimized wavefunction
x_pigs, y_pigs, err_pigs = np.loadtxt("8.3/data/variational/probability.dat", unpack = True)
axes[1].errorbar(x_pigs, y_pigs, yerr = err_pigs, label = "PIGS")
axes[1].plot(x_sample,y_sample, label="VMC code")
axes[1].set_title("PIGS - Variationally optimized wavefunction")


for ax in axes:
    ax.set_ylabel(r"$|\psi_0(x)|^2$")
    ax.set_xlabel(r"$x$")
    ax.plot(x, (psi[0])**2, linestyle='dashed', color='r', label = "Numerical solution")
    ax.set_xlim((-3, 3))
    ax.legend()
    ax.grid(True)

fig.tight_layout()
plt.show()
```

The results obtaines with the PIGS code are more accurate than the Variational montecarlo previously used.
Using the optimized wavefunction, the imaginary time needed to obtain good results can be lower, as we expect.
The wavefunction $\Psi_T^{\sigma,\mu}(x)$ is a better approximation of the actual ground state, so the convergence is faster.

The imaginary time propagation for the constant wavefunction was set to 3.5, while using the optimized one it can be set at 2.0. Using lower imaginary times negatively affect the results, that starts detaching from the numerical solution.

Then, the temperature is increased, and the variation in the sampled configuration histogram are shown in the final plot.
