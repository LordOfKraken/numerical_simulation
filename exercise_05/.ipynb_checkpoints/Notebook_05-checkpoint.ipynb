The sampling of $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ is performed in Cartesian coordinates using a uniform distribution and a normal distribution.
For each distribution, the wave function is sampled starting from two different starting point, one close and one far from the origin.

The software does not read an input file, but executes the step adjustment, equilibration and data blocking sequentially.

### Step adjustment

First, the metropolis step is set to have an acceptance ratio $\sim$ 50%. This is achieved testing various steps length in the range 0.5 / 5 for the uniform distribution, and 0 / 5 for the gaussian sigma , using a bisection algorithm to change the step until the desired acceptance ratio is reached.

The test, consist in tryng to evolve the system 10000 times, saving the final acceptance ratio for each sampling distribution (uniform and normal) and each state of the wavefunction (1s and 2p), so 4 different steps are used.

### Equilibration

The two states are sampled for 2000 steps, using two starting points, one close and one far from the origin, to study the equilibration steps needed.

The closer point is at (1,0,0), while the far one is at (100,100,100). The coordinates and the distance shown in all the plots are in a0 units.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 5))
axes = axes.flatten()

radius_raw = np.loadtxt("5.1/data/r_close.out")
axes[0].plot(radius_raw)
axes[0].set_title("Sampled radius starting from (1,0,0)")

radius_raw = np.loadtxt("5.1/data/r_far.out")
axes[1].plot(radius_raw)
axes[1].set_title("Sampled radius starting from (100,100,100)")

for ax in axes:
    ax.set_xlabel("Steps")
    ax.set_ylabel("Radius")
    ax.grid(True)
    ax.legend(["1s uniform", "1s normal", "2p uniform", "2p normal"])

fig.tight_layout()
plt.show()
```

The plot shows that if the starting point is far from the origin, some steps of equilibration are needed before the simulation can start. The right side plot shows that 1000 steps should be enough to equilibrate the system. 

### Block size

The previous sampling, with the system now equilibrated, is continued for another 1000 steps, and the optimal block size can be chosen studying the autocorrelation factor of the data.

```python
def autocovariance(series, lag):
    n = len(series)
    mean = 0
    for x in series:
        mean += x
    mean /= n
    c = 0
    for i in range(n - lag):
        c += (series[i] - mean)*(series[i + lag] - mean)
    return c / n

def autocorrelation(series, lag):
    return autocovariance(series, lag) / autocovariance(series, 0)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 5))
axes = axes.flatten()
lag = 500
    
r_close = np.loadtxt("5.1/data/r_close_eq.out",unpack=True)

for dataset in r_close:
    axes[0].plot([autocorrelation(dataset, l) for l in range(lag)])
axes[0].set_title("Autocorrelation function starting close")

r_far = np.loadtxt("5.1/data/r_far_eq.out", unpack = True)

for dataset in r_far:
    axes[1].plot([autocorrelation(dataset, l) for l in range(lag)])
axes[1].set_title("Autocorrelation function starting far")

for ax in axes:
    ax.grid(True)
    ax.set_xlabel("Lag")
    ax.set_ylabel("Autocorrelation")
    ax.legend(["1s uniform", "1s normal", "2p uniform", "2p normal"])

fig.tight_layout()
plt.show()
```

There are some fluctuations in the autocorrelation, even with high lag time, that may be due to imperfection in the rangom number generator. However, it's still possible to see the expected exponential decay $e^{-t/\tau}$, so I try to fit the data to find the autocorrelation length $\tau$.

```python
from scipy import optimize

def exp_fit(t, tau):
    return np.exp(-t / tau)
lag = 500

print("Autocorrelation length starting close:")
for dataset in r_close:
    corr = [autocorrelation(dataset, l) for l in range(lag)]
    tau, err_tau = optimize.curve_fit(exp_fit, np.arange(lag), corr)
    print(str(tau[0]) + " +- " + str(err_tau[0][0]))

    
print("Autocorrelation length starting far")
for dataset in r_far:
    corr = [autocorrelation(dataset, l) for l in range(lag)]
    tau, err_tau = optimize.curve_fit(exp_fit, np.arange(lag), corr)
    print(str(tau[0]) + " +- " + str(err_tau[0][0]))
```

The highest autocorrelation length is $\sim 15$ so the minimal block size should be $\sim 30$, but since the fluctuations are very high, I choose a block size of 50.

### Radius evaluation

Finally, using the parameters found the previous analysis, it's possible to draw an estimate of $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ with the respective uncertainties, using the data blocking method with M = $10^6$ throws, and a block size of 50. 
Since the system now it's equilibrated, i dont make two different simulations starting close and far, but just one.

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 5))
axes = axes.flatten()

r_1s = np.loadtxt("5.1/data/r_av_1s.out", unpack = True)
r_1s_g = np.loadtxt("5.1/data/r_av_1s_g.out", unpack = True)
x = range(len(r_1s[0]))

axes[0].errorbar(x, r_1s[0], yerr = r_1s[1], label = "Uniform step",elinewidth = 1)
axes[0].errorbar(x, r_1s_g[0], yerr = r_1s_g[1], label = "Normal step", elinewidth = 1)
axes[0].hlines(3/2,0,len(r_1s[0])-1,colors = "red", linestyle = "dashed", label = "Expected value")  

axes[0].set_title("Average radius of state 1s")
axes[0].set_ylabel(r"$\langle r\rangle_{1,0,0}$")
axes[0].set_xlabel("Block number")
axes[0].legend()
axes[0].grid(True)

r_2p = np.loadtxt("5.1/data/r_av_2p.out", unpack = True)
r_2p_g = np.loadtxt("5.1/data/r_av_2p_g.out", unpack = True)
x = range(len(r_2p[0]))

axes[1].errorbar(x, r_2p[0], yerr = r_2p[1], label = "Uniform step")
axes[1].errorbar(x, r_2p_g[0], yerr = r_2p_g[1], label = "Normal step")
axes[1].hlines(5,0,len(r_2p[0])-1,colors = "red", linestyle = "dashed", label = "Expected value")  

axes[1].set_title("Average radius of state 2p")
axes[1].set_ylabel(r"$\langle r\rangle_{2,1,0}$")
axes[1].set_xlabel("Block number")
axes[1].legend()
axes[1].grid(True)

fig.tight_layout()
plt.show()
```
