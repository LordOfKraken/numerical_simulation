### Exercise 03.1
- Compute at time $t=0$ via Monte Carlo the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$, using the following parameters: 

1. asset price at $t=0$: $S(0)=100$
2. delivery time: $T=1$
3. strike price: $K=100$
4. risk-free interest rate: $r=0.1$
5. volatility: $\sigma=0.25$

For the blocking method, i have used $10^4$ throws, and $10^2$ blocks

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

#### Direct sampling


```python
x, y, err_prog = np.loadtxt("3.1/data/call_price_direct.out", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
plt.figure(figsize=(14,7))
plt.errorbar(x,y,yerr=err_prog,label='Call direct')

# analytic solution for comparison
plt.hlines(14.975790778311286, 0, 100, colors='r', linestyles='dashed', label='Analytic solution')
plt.title('direct Call-option price')
plt.xlabel('# block')
plt.ylabel('Call price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x, y, err_prog = np.loadtxt("3.1/data/put_price_direct.out", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
plt.figure(figsize=(14,7))
plt.errorbar(x,y,yerr=err_prog,label='Put direct')

# analytic solution for comparison
plt.hlines(5.4595325819072364, 0, 100, colors='r', linestyles='dashed', label = 'Analytic solution')
plt.title('direct Put-option price')
plt.xlabel('# block')
plt.ylabel('Put price')
plt.legend()
plt.grid(True)
plt.show()
```
