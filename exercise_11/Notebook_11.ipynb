### Exercise 1

Fit the following linear function
$$
f(x)=2x+1
$$

Studyng the result with different epochs ($N_{epochs}$), data points ($N_{train}$) and data noise ($\sigma$).

```python
#modules
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

#from mpl_toolkits.mplot3d import Axes3D

#random setup
np.random.seed(0)

#function parameters
m = 2 # slope
b = 1 # intersect

# generate starting data
x_train = np.random.uniform(-1, 1, 10000)
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function
```

```python
#NN setup, single neuron
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python

# parameters range to test
epochs = [10, 50,100]
data = [100, 500, 1000]
sigma = [0.1, 0.2, 0.3]

fig, ax = plt.subplots(nrows = len(sigma)*len(epochs), ncols = 2, figsize = (6*2, len(sigma)*len(epochs) * 5))
ax = ax.flatten()


for s in range(len(sigma)):
    for e in range(len(epochs)):
        for d in range(len(data)):
            # generate data with a sigma
            y_train = np.random.normal(m * x_train + b, sigma[s]) 
            y_valid = np.random.normal(m * x_valid + b, sigma[s])
            
            #train with the current parameters
            history = model.fit(x=x_train[:data[d]], y=y_train[:data[d]], batch_size=32, epochs=epochs[e],
              shuffle=True, validation_data=(x_valid, y_valid),verbose=0)
            
            #generate predicted data
            x_pred = np.random.uniform(-1, 1, 100)
            y_pred = model.predict(x_pred)
            
            # fill the loss function plots
            ax[s*2*len(epochs) + 2*e].plot(history.history['loss'],label= "Train " + str(data[d]) + " points")
            ax[s*2*len(epochs) + 2*e].plot(history.history['val_loss'],label= "Val " + str(data[d]) + " points")
            
            #fill the fit plot
            ax[s*2*len(epochs) + 2*e + 1].scatter(x_pred, y_pred, s=10, label="Prediction " + str(data[d]) + " points")
            
        
        #set titles, etc
        ax[s*2*len(epochs) + 2*e].set_title("Loss function\n" + str(sigma[s]) + " noise, " + 
                                           str(epochs[e]) + " epochs" )
        ax[s*2*len(epochs) + 2*e].set_xlabel("Epochs")
        ax[s*2*len(epochs) + 2*e].set_ylabel("Loss value")
        
        ax[s*2*len(epochs) + 2*e + 1].set_title("Predicted points\n" + str(sigma[s]) + " noise, " + 
                                           str(epochs[e]) + " epochs")
        ax[s*2*len(epochs) + 2*e + 1].set_xlabel("X")
        ax[s*2*len(epochs) + 2*e + 1].set_ylabel("Y")
        ax[s*2*len(epochs) + 2*e + 1].plot(x_valid, y_target, label="Target")

        
for a in ax:
    a.legend()
    a.grid(True)

fig.tight_layout(pad=3.0)
plt.show()
        
        
```

As expected, the best results are obtained with bigger datasets, and more epochs, with the number of training data being the most important parameter.

A note to read the plots: the validation dataset has the same size (100 points) for every plot, the labes are referred to the size of the training dataset, and Val is the `val_loss` value.

### Execise 2

Extend the model to fit the polynomial

$$
f(x)=4-3x-2x^2+3x^3, \qquad x \in [-1,1]
$$

exploring reasonable different choices for:

- the number of layers
- the number of neurons in each layer
- the activation function
- the optimizer
- the loss function

At first, i tried changing the structure of the NN itself, as number of layers and neurons per layer.

```python
def func(x):
    return 4. - 3.*x -2.*x**2 + 3.*x**3

def set_model(n_layers, neurons_per_layer, activ_fn, optim, loss_fn):
    
    model = keras.models.Sequential()
    initializer = tf.keras.initializers.GlorotUniform(seed = 0)
    # input layer, which specify shape
    model.add(keras.layers.Dense(neurons_per_layer, activation=activ_fn,kernel_initializer=initializer, input_shape=(1,)))
    
    # internal (deep) layers, the n_layer refers only to these, not counting input and output
    for i in range(1, n_layers-1):
        model.add(keras.layers.Dense(neurons_per_layer, activation=activ_fn))
        
    # output layer, with 1 neuron
    model.add(keras.layers.Dense(1, activation=activ_fn))
    
    model.compile(loss=loss_fn, optimizer=optim, metrics=['mse'])
    return model
```

```python
#layers=[1,2,5,10,20]
#n_per_layer=[2,5,10,20,30]

layers=[1,2,5,10,20]
n_per_layer=[5,10,20,30]

np.random.seed(19997)
tf.random.set_seed(19997)

activ_fn='relu'
optim='sgd'
loss_fn='mse'

# training data
x_train = np.random.uniform(-1, 1, 200)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()

sigma=0.1

y_target = func(x_valid)
y_train = np.random.normal(func(x_train), sigma)
y_valid = np.random.normal(func(x_valid), sigma)


#y_train = np.random.normal(4. - 3.*x_train - 2.*x_train**2 + 3.*x_train**3, sigma)
#y_valid = np.random.normal(4. - 3.*x_valid - 2.*x_valid**2 + 3.*x_valid**3, sigma)
```

```python
line = 1

for l in layers:
    for neuron in n_per_layer:
        
        plt.figure(line, figsize = (15, 4))
        K.clear_session()
              
        model = set_model(l, neuron, activ_fn, optim, loss_fn)
        history = model.fit(x=x_train, y=y_train, epochs=100 ,shuffle=True, 
                            validation_data=(x_valid, y_valid),verbose=False)
        
        #print( "Layers = " + str(l) + ", Neurons per layer = " + str(neuron) + 
        #       ", Final loss = " + str(history.history['val_loss'][-1]))

        # loss value
        plt.subplot(1, 3, 1)
        plt.plot(history.history['loss'],label="Train")
        plt.plot(history.history['val_loss'],label="Validation")
        plt.title('Model loss')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)
  
        
        # fit inside training range   
        plt.subplot(1, 3, 2)
        x_pred = np.random.uniform(-1, 1, 100)
        y_pred = model.predict(x_pred)
        
        plt.scatter(x_pred, y_pred, label="Prediction")
        plt.plot(x_valid, y_target, color='r', label="Target")
        plt.title("Layers = " + str(l) + ", Neurons per layer = " + str(neuron) + 
               ", Final loss = "  + str(history.history['val_loss'][-1]) + 
                '\n\nPredictions inside training range')
        plt.ylabel('Y')
        plt.xlabel('X')
        plt.grid(True)
        plt.legend()

  
        # fit outside range
        plt.subplot(1, 3, 3)
        x_over = np.random.uniform(-2, 2, 200)
        x_over.sort()
        y_over_target = func(x_over)
        y_pred = model.predict(x_over)
        
        plt.scatter(x_over, y_pred, label="Prediction")
        plt.plot(x_over, y_over_target,color='r', label="Target")
        plt.title('Predictions outside training range')
        plt.ylabel('Y')
        plt.xlabel('X')
        plt.grid(True)
        plt.legend()
        
        K.clear_session()
        
        line +=1
        plt.show()


    
```

As expected, no model it's able to predict data outside the training data range. The best results are obtained using more neurons, and more layers, although the number of neurons per layer seems to be the most important parameter.

Multiple layers with very few neurons have worse permormance than few layer with many neurons each, but using too many neurons results in overfitting, and again, worse results.

At this point, i try to change the other parameters, like the activation function, the optimizer and the loss function, for a single model with 5 layers and 20 neuron per layer


```python
layers=5
neurons_per_layer=20
activation=['relu', 'elu']
optimizers=['sgd', 'adam']
loss=['mean_squared_error', 'mean_absolute_error']

# training data
x_train = np.random.uniform(-1, 1, 200)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = func(x_valid)

sigma=0.1
y_train = np.random.normal(func(x_train), sigma)
y_valid = np.random.normal(func(x_valid), sigma)

line=1

for act in activation:
    for opt in optimizers:
        for l in loss:
        
            plt.figure(line, figsize = (15, 4))
            K.clear_session()
            model = set_model(layers, neurons_per_layer, act, opt, l)
            history = model.fit(x=x_train, y=y_train, batch_size=32, epochs=100 ,
                                shuffle=True, validation_data=(x_valid, y_valid), verbose=0)


            # fill the plots
            # loss value
            plt.subplot(1,3,1)
            plt.plot(history.history['loss'],label="Train")
            plt.plot(history.history['val_loss'], label = "Validation")
            plt.title('Model loss')
            plt.ylabel('Loss')
            plt.xlabel('Epoch')
            plt.legend()
            plt.grid(True)

            # fit inside range
            plt.subplot(1,3,2)
            x_pred = np.random.uniform(-1, 1, 100)
            y_pred = model.predict(x_pred)
            plt.scatter(x_pred, y_pred, label="Prediction")
            plt.plot(x_valid, y_target, color='r', label="Target")
            plt.title("Activation function = " + str(act) + 
                ", Optimizer = " + str(opt) +
                ", Loss function = " + str(l) + 
                ", Final loss = "  + str(history.history['val_loss'][-1]) + 
                '\n\nPredictions inside training range')
            plt.ylabel('Y')
            plt.xlabel('X')
            plt.legend()
            plt.grid(True)

            # fit outside range
            plt.subplot(1,3,3)
            x_over = np.random.uniform(-2, 2, 200)
            x_over.sort()
            y_over_target = func(x_over)
            y_pred = model.predict(x_over)

            plt.scatter(x_over, y_pred, label="Prediction")
            plt.plot(x_over, y_over_target,color='r', label="Target")
            plt.title('Predictions outside training range')
            plt.ylabel('Y')
            plt.xlabel('X')
            plt.legend()
            plt.grid(True)
            
            line+=1
            K.clear_session()
            plt.show()
```

The best results for this Neural network are obtained with the relu activation function, the adam optimizer, and the mean squared error as loss function

### Exercise 3

In the end, I try to extend the model to fit a simple trigonometric 2D function 
$$f(x,y) = \sin(x^2+y^2)$$
in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
np.random.seed(0)

x_train = np.random.uniform(-1.5, 1.5, 1000)
y_train =np.random.uniform(-1.5, 1.5, 1000)
x_valid = np.random.uniform(-1.5, 1.5, 500)
y_valid = np.random.uniform(-1.5, 1.5, 500)

z_target = np.sin(x_valid**2+y_valid**2)

sigma=0.1
z_train = np.random.normal(np.sin(x_train**2+y_train**2), sigma)
z_valid = np.random.normal(np.sin(x_valid**2+y_valid**2), sigma)

data_train=np.array([x_train, y_train]).T
data_valid=np.array([x_valid, y_valid]).T


# redefine for input shape 2
def set_model(n_layers, neurons_per_layer, activ_fn, optim, loss_fn):
    
    model = keras.models.Sequential()
    # input layer, which specify shape
    model.add(keras.layers.Dense(neurons_per_layer, activation=activ_fn, input_shape=(2,)))
    
    # internal layers, the n_layer refers only to these, not counting input and output
    for i in range(0, n_layers):
        model.add(keras.layers.Dense(neurons_per_layer, activation=activ_fn))
        
    # output layer, with 1 neuron
    model.add(keras.layers.Dense(1, activation=activ_fn))
    model.compile(loss=loss_fn, optimizer=optim, metrics=['mse'])
    return model
```

```python
layers=10
neurons_per_layer=20 # more neurons are needed for a more complex function
activation='relu'
optimizer='sgd'
loss='mse'

np.random.seed(0)


model = set_model(layers, neurons_per_layer, activation, optimizer, loss)

history = model.fit(x=data_train, y=z_train, batch_size=32, epochs=100 ,shuffle=True, validation_data=(data_valid, z_valid), verbose=0)

plt.figure(figsize = (9, 6))
plt.plot(history.history['loss'],label = 'Train')
plt.plot(history.history['val_loss'],label="Validation")
plt.title('Loss value')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
```

```python
fig = plt.figure(figsize = (9, 6))
ax = fig.gca(projection='3d')

ax.plot_trisurf(x_valid, y_valid, z_target, linewidth=0.2, antialiased=True)
x_predicted = np.random.uniform(-1.5, 1.5, 500)
y_predicted = np.random.uniform(-1.5, 1.5, 500)
data_predicted = np.array([x_predicted, y_predicted]).T
z_predicted = model.predict(data_predicted)
ax.scatter(x_predicted, y_predicted, z_predicted, color='r', s=3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

The point predicted seems to have a good match with the actual function, although the loss function doesn't seem to be stable yet, so a longer training time may improve the results
